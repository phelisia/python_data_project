```python
from datasets import load_dataset
from matplotlib import pyplot as plt
import pandas as pd
dataset=load_dataset('lukebarousse/data_jobs') 
```

```python
df=dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
```

```python
df
```

```python
df_USA_software_pivot=pd.read_csv('https://lukeb.co/software_csv')
df_USA_software_pivot
```

```python
df_usa=df[df['job_country']=='United States'].copy()
df_usa
```

```python
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_USA=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_USA
```

```python
df_USA.reset_index(inplace=True)
df_USA['month_no']=pd.to_datetime(df_USA['job_posted_month'],format='%B').dt.month
df_USA.sort_values(by='month_no',inplace=True)
df_USA.set_index('job_posted_month',inplace=True)
df_USA.drop(columns='month_no',inplace=True)
df_USA
```

```python
df_US_Pivot=df_USA
```

```python
df_US_Pivot
```

```python
df_US_merged=df_US_Pivot.merge(df_USA_software_pivot,on='job_posted_month')
```

```python
df_US_merged
```

```python
top_5=df_US_merged.sum().sort_values(ascending=False).head().index.to_list()
top_5
```

```python
df_US_merged[top_5].plot(kind='line')
```
