```python
from datasets import load_dataset
from matplotlib import pyplot as plt
import pandas as pd
dataset=load_dataset('lukebarousse/data_jobs') 
```

```python
df=dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
```

```python
import ast
def clean_list(skill_list):
    if pd.notna(skill_list):
       return ast.literal_eval(skill_list)
df['job_skills']=df['job_skills'].apply(clean_list)
```

```python
df=df[df['job_title_short']=='Data Analyst']
```

```python
df
```

```python
df_explode=df.explode('job_skills')
df_explode
```

```python
skill_stats=df_explode.groupby('job_skills').agg(
    skills_count=('job_skills','count'),
    median_salary=('salary_year_avg','median')
)
```

```python
skill_stats=skill_stats.sort_values(by='skills_count',ascending=False).head(10)
skill_stats
```

```python

from adjustText import adjust_text
skill_stats.plot(kind='scatter',x='skills_count',y='median_salary')
texts=[]
for i, txt in enumerate(skill_stats.index):
    texts.append(plt.text(skill_stats['skills_count'].iloc[i],skill_stats['median_salary'].iloc[i],txt))
adjust_text(texts)
ax=plt.gca()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y,pos:f'${int (y/1000)}k'))
```
