```python
from datasets import load_dataset
from matplotlib import pyplot as plt
import pandas as pd
dataset=load_dataset('lukebarousse/data_jobs') 
```

```python
df=dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df.job_posted_date)
```

```python
import ast
def clean_list(skill_list):
    if pd.notna(skill_list):
       return ast.literal_eval(skill_list)
df['job_skills']=df['job_skills'].apply(clean_list)
```

```python
df
```

```python
job_titles=['Data Analyst', 'Data Scientist','Data Engineer']
df_DA_USA = df[(df['job_title_short'].isin(job_titles)) & (df['job_country'] == 'United States')].copy()
df_DA_USA
```

```python
df_DA_USA['salary_year_avg'].plot(kind='hist',bins=40,edgecolor='black')
plt.xlim(0,250000)
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x,pos:f'${int (x/1000)}k'))
```

```python
# Ensure df_DA_USA is a DataFrame and not a Series

df_DA_USA = df_DA_USA.dropna(subset=['salary_year_avg'])

# Step 3: Sample 10 rows from the filtered DataFrame
# df_DA_USA = df_DA_USA.sample(10)

# Step 4: Plot the 'salary_year_avg' column using a box plot
df_DA_USA['salary_year_avg'].plot(kind='box', vert=False)

# Display the plot
plt.title('Box Plot of Sampled Data Analyst Salaries in the United States')
plt.xlabel('Salary (Yearly Average)')
plt.show()
```

```python
job_list = [df_DA_USA['salary_year_avg'][df_DA_USA['job_title_short'] == job_title] for job_title in job_titles]
```

```python
plt.boxplot(job_list,labels=job_titles,vert=False)
```
